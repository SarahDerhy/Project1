```python
fig, axes = plt.subplots(figsize=(10, 4), ncols=2)

sns.scatterplot(data = student_df, y = 'final_grade', x = 'AtLeastTwoYes', ax = axes[0])
axes[0].set_yscale('linear')

sns.stripplot(data = student_df, y = 'final_grade', x = "No_extra_support", ax = axes[1])
axes[1].set_yscale('linear')

plt.tight_layout()
plt.show()
```

```python


# Création de la colonne 'AtLeastTwoYes'
student_df['AtLeastTwoYes'] = student_df[['extra_paid_classes', 'school_support', 'internet_access']].apply(lambda x: 'Yes' if (x == 'yes').sum() >= 2 else 'No', axis=1)

# Tracé du diagramme en boîte
plt.figure(figsize=(10, 6))
boxplot = student_df.boxplot(column='final_grade', by='AtLeastTwoYes')
plt.grid(False)
plt.title('Influence d\'avoir au moins deux "yes" sur la note des élèves')
plt.suptitle('')
plt.xlabel('Au moins deux "yes"')
plt.ylabel('Note')
plt.show()
```

in this boxplot, we tried to see if a student wich has at least 2 additional learning supports  will have a higher final grade.

```python
study_time_order = ["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"]

# Trier le DataFrame en fonction de la colonne 'study_time' dans l'ordre spécifié
sorted_df = student_df.sort_values(by="study_time", key=lambda x: x.map({val: i for i, val in enumerate(study_time_order)}))

# Tracer le graphique en ligne en utilisant le DataFrame trié
sns.lineplot(data=sorted_df, x="study_time", y="final_grade",sort=False)
plt.ylim(0,20)
           
           
           
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

study_time_order = ["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"]

fig, axes = plt.subplots(figsize=(12, 4), ncols=2)

# Tracer le graphique en nuage de points
sns.scatterplot(x="study_time", y="final_grade", data=student_df, ax=axes[1])

# Tracer le graphique en barres
sns.barplot(x="study_time", y="final_grade", data=student_df, ax=axes[0])

# Réglage des limites de l'axe y pour les deux graphiques
axes[0].set_ylim(0, 20)
axes[1].set_ylim(0, 20)

# Définir les marqueurs d'axe personnalisés pour les graphiques en nuage de points
axes[1].set_xticks(range(len(study_time_order)))
axes[1].set_xticklabels(study_time_order)

plt.show()
```
